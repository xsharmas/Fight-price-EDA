```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

to handle this null we can replace this null value with any relevent catagory change we are going to do

```python
df['Total_Stops'] = df['Total_Stops'].map({
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4
})
df['Total_Stops'] = df['Total_Stops'].fillna(1)  # Fill missing with 1 stop
df['Total_Stops'] = df['Total_Stops'].astype(int)
```
